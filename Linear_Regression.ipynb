#1 Single variable Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#  First case
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='red', marker="+")
```

```python
# using linear regression

reg = linear_model.LinearRegression()
X = df[['area']]
y = df.price
reg.fit(X, y)
```

```python
# value of m
m = reg.coef_
m
```

```python
# value of b
b = reg.intercept_
b
```

```python
reg.predict([[5000]])
x = 5000
```

```python
y = m*x+b
y
```

```python
%matplotlib inline
plt.xlabel('area', fontsize = 20)
plt.ylabel('area', fontsize = 20)
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
```

```python
# Second Case

d = pd.read_csv("areas.csv")
d
```

```python
p = reg.predict(d)
p
```

```python
d['prices'] = p
d
```

```python
d.to_csv("prediction.csv", index=False)
```

#2 Multivariable Linear Regression

```python
# url1 = "https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv"
df2 = pd.read_csv("homeprices multi.csv")
df2
```

```python
import math
median_bedrooms = math.floor(df2.bedrooms.median()) # using math.floor for integers
median_bedrooms
```

```python
df2.bedrooms.fillna(median_bedrooms, inplace=True)
# df2['bedrooms'][2] = 3
df2
```

```python
reg = linear_model.LinearRegression()
reg.fit(df2[['area','bedrooms','age']],df2.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3, 40]])
```
