#1. Stationarity

##1.1 ADF

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(sales['Net Sales'])

print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print(f'   {key}: {value}')
```


The ADF test result indicates that the 'Net Sales' data is already stationary without requiring differencing. The very low p-value (0.0) and ADF statistic (-21.63) significantly below critical values confirm stationarity. Therefore, differencing may not be necessary for this dataset since it's already in a stationary form.

#Data Transformation

```python
df['Net Sales Diff'] = df['Net Sales'].diff()
```

```python
positive_sales = df[df['Net Sales'] > 0]['Net Sales']


df['Net Sales Log'] = np.log(positive_sales)
```

```python
# Assuming you chose differencing
df['Net Sales Diff'] = df['Net Sales'].diff().dropna()

# Perform ADF test on the differenced data
result_diff = adfuller(df['Net Sales Diff'].dropna())

# Print the ADF test results for the differenced data
print(f'ADF Statistic: {result_diff[0]}')
print(f'p-value: {result_diff[1]}')
print(f'Num of lags : {result_diff[2]}')
print(f'Num of observations used for ADF Regression and critical values calculation: {result_diff[3]}')
print('Critical Values:')
# print(f'')
for key, value in result_diff[4].items():
    print(f'   {key}: {value}')

# Interpret the results
if result_diff[1] <= 0.05:
    print("The differenced time series data is likely stationary.")
else:
    print("The differenced time series data is likely non-stationary.")
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot autocorrelation and partial autocorrelation
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['Net Sales Diff'].dropna(), lags=40, ax=ax1)
plot_pacf(df['Net Sales Diff'].dropna(), lags=40, ax=ax2)
plt.show()
```

it displays strong short-term dependence at lag 1, indicating a notable correlation between consecutive time points. Additionally, smaller peaks at lags 5 and 10 suggest some longer-term dependence but not as pronounced as the short-term correlation. Absence of clear seasonal patterns in the ACF plot implies no predictable seasonal trends. The PACF plot's lack of significant peaks suggests potential stationarity, indicating consistent statistical properties over time.

```python
import itertools
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")


p_values = range(0, 3)
d_values = range(0, 2)
q_values = range(0, 3)


parameter_combinations = list(itertools.product(p_values, d_values, q_values))

best_aic = float("inf")
best_params = None

for param in parameter_combinations:
    try:
        model = ARIMA(df_copy['Net Sales'], order=param)
        fitted_model = model.fit()
        aic = fitted_model.aic

        if aic < best_aic:
            best_aic = aic
            best_params = param
    except:
        continue

print(f"Best AIC: {best_aic}")
print(f"Best Parameters (p, d, q): {best_params}")
```

```python
best_model = ARIMA(df_copy['Net Sales_log'], order=best_params)
fitted_best_model = best_model.fit()

# Assess model performance
print(fitted_best_model.summary())
```

```python
# Analyze residuals
residuals = fitted_best_model.resid


from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(residuals, ax=ax[0])
plot_pacf(residuals, ax=ax[1])
plt.show()






```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA


unique_drugs = df['Stock_Name Encoded'].unique()
unique_areas = df['Area.C'].unique()

forecasted_data = []

for drug in unique_drugs:
    for area in unique_areas:
        drug_area_data = df[(df['Stock_Name Encoded'] == drug) & (df['Area.C'] == area)]['Net Sales']

        best_aic = float("inf")
        best_params = None

        for param in parameter_combinations:
            try:
                model = ARIMA(drug_area_data, order=param)
                fitted_model = model.fit()
                aic = fitted_model.aic


                if aic < best_aic:
                    best_aic = aic
                    best_params = param
            except:
                continue

        if best_params is None:
            continue


        print(f"Drug: {drug}, Area Code: {area}")
        print(f"Best AIC: {best_aic}")
        print(f"Best Parameters (p, d, q): {best_params}")

        final_model = ARIMA(drug_area_data, order=best_params)
        final_fitted_model = final_model.fit()


        forecast = final_fitted_model.forecast(steps=10)  #
        print("Forecasted values:")
        print(forecast)


        forecast_df = pd.DataFrame({
             'Year': range(2021, 2031),
             'Net Sales': forecast,
             'Area.C': [area] * 10,
             'Stock_Name Encoded': [drug] * 10
        })

        forecasted_data.append(forecast_df)


forecast_combined = pd.concat(forecasted_data, ignore_index=True)

forecast_combined.to_csv('forecast.csv', index=False)
```
